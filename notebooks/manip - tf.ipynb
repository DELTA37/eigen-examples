```python
import tensorflow as tf
import numpy as np
```

# manip has one differential operation

```python
tf.reset_default_graph()
x = tf.placeholder_with_default(input=tf.constant([1, 2, 3], dtype=tf.float32), name='x', shape=[3])
y = tf.placeholder(dtype=tf.float32, shape=[])
a = tf.get_variable('a', shape=[3])
y_pred = tf.reduce_sum(a * tf.manip.roll(x, axis=0, shift=2))
loss = (y - y_pred)**2
train_step = tf.train.AdamOptimizer(learning_rate=0.01).minimize(loss)
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```

```python
for i in np.linspace(-10, 10, 10000):
    sess.run(train_step, feed_dict={x : [i, np.random.rand(), np.random.rand()], y : i})
    print(sess.run([a, loss], feed_dict={x : [i, np.random.rand(), np.random.rand()], y : i}))
```
